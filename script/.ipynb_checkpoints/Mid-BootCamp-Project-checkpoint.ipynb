#### Get Benchmark Result

```python
df['zipcode'] = df['zipcode'].astype('int64')
```

```python
# Spliting X - y data
X = df.drop('price', axis=1)
y = df['price']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
prediction = model.predict(X_test)
```

```python
r2_score(y_test, prediction), mean_absolute_error(y_test, prediction), mean_squared_error(y_test, prediction, squared=False)
```

### Feature Engineering

```python
#Drop outliers above 650k from Price
df2 = df.copy()
df2 = df2[df2['price'] <= 650000]
df2
```

```python
df2['zipcode'] = df2['zipcode'].astype('category')
```

```python
# Spliting X - y data
X = df2.drop('price', axis=1)
y = df2['price']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
#Train data
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
#Test trained data
pred = model.predict(X_test)
```

```python
r2_score(y_test, pred), mean_absolute_error(y_test, pred), mean_squared_error(y_test, pred, squared=False), math.sqrt(mean_squared_error(y_test, pred, squared=False)) 
```

I removed the house prices above 650k(outliers) from Price because we will still have another model for houses above 650k.

Removal of these high prices didn't quite improve my R-squared value. But it reduced the error of my model significantly which I consider important.
